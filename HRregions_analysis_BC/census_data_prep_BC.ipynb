```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# Import the census data
df = pd.read_csv('98-401-X2016058_English_CSV_data.csv')
df.info()
df.head()
```

```python
# Show that BCs data is aggregated at the HSA level only
df['GEO_NAME'].value_counts()
len(df[df ['GEO_NAME'].str.contains('Interior', regex=False, na=False)])
```

```python
# Filter for BC HSA/HR data
listofHSA = [5942, 5911, 5921, 5922, 5923, 5912, 
             5933, 5943, 5953, 5952, 5951, 5913, 
             5931, 5941, 5914, 5932]

df = df[df['GEO_CODE (POR)'].isin(listofHSA)]
df.info()
df['GEO_CODE (POR)'].nunique()
df.head()
```

```python
df['GEO_NAME'].value_counts()
```

```python
# Import list of variables of interest
char = pd.read_csv('selected_variables.csv', header=None)
char.info()

char[1] = char[1].map(lambda x: x.lstrip(' '))
char[1]
```

```python
# Subset the data
listofid = char[0].to_list()
subset = df.loc[df['Member ID: Profile of Health Regions (2247)'].isin(listofid)]  
subset = subset.reset_index()
subset.info()
subset.head()
```

```python
# Drop columns that are not needed
listofcols = ['index', 'CENSUS_YEAR', 'GEO_LEVEL', 'GNR','GNR_LF', 'DATA_QUALITY_FLAG', 'ALT_GEO_CODE', 
              'Dim: Sex (3): Member ID: [2]: Male', 'Dim: Sex (3): Member ID: [3]: Female']

subset.drop(listofcols, axis = 1, inplace = True)
subset["Dim: Sex (3): Member ID: [1]: Total - Sex"] = pd.to_numeric(subset["Dim: Sex (3): Member ID: [1]: Total - Sex"])
subset.head()
```

```python
listofcols = list(subset['DIM: Profile of Health Regions (2247)'].head(35))
listofcols
```

```python
# Pivot the table
table = pd.pivot_table(subset, values = 'Dim: Sex (3): Member ID: [1]: Total - Sex',
                       index = ['GEO_NAME', 'GEO_CODE (POR)'],
                       columns = ['DIM: Profile of Health Regions (2247)'])
table.info()
table.head()
table = table.reset_index()
```

```python
table.head()
# table.columns.values
```

```python
# Reorder the columns
listofcols = ['GEO_NAME', 'GEO_CODE (POR)'] + listofcols

table = table.reindex(columns=listofcols)
table.head()
```

```python
# Create new columns with proportions for the count data
table['Single-detached house prop'] = table['Single-detached house']/table['Total - Occupied private dwellings by structural type of dwelling - 100% data']
table['Married or living common law prop'] = table['Married or living common law']/table['Total - Marital status for the population aged 15 years and over - 100% data']
table['Not married and not living common law prop'] = table['Not married and not living common law']/table['Total - Marital status for the population aged 15 years and over - 100% data']

table['Owner prop'] = table ['Owner']/table['Total - Private households by tenure - 25% sample data']
table['Renter prop'] = table ['Renter']/table['Total - Private households by tenure - 25% sample data']

table['One person or fewer per room prop'] = table ['One person or fewer per room']/table['Total - Private households by number of persons per room - 25% sample data']
table['More than 1 person per room prop'] = table ['More than 1 person per room']/table['Total - Private households by number of persons per room - 25% sample data']

table['Suitable prop'] = table ['Suitable']/table['Total - Private households by housing suitability - 25% sample data']
table['Not suitable prop'] = table ['Not suitable']/table['Total - Private households by housing suitability - 25% sample data']

table['No certificate, diploma or degree prop'] = table ['No certificate, diploma or degree']/table['Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data']
table['Secondary (high) school diploma or equivalency certificate prop'] = table ['Secondary (high) school diploma or equivalency certificate']/table['Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data']
table['Postsecondary certificate, diploma or degree prop'] = table ['Postsecondary certificate, diploma or degree']/table['Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data']

table['In the labour force prop'] = table ['In the labour force']/table['Total - Population aged 15 years and over by Labour force status - 25% sample data']
table['Employed prop'] = table ['Employed']/table['In the labour force']
table['Unemployed prop'] = table ['Unemployed']/table['In the labour force']
table['Not in the labour force prop'] = table ['Not in the labour force']/table['Total - Population aged 15 years and over by Labour force status - 25% sample data']

table['Car, truck, van - as a driver prop'] = table ['Car, truck, van - as a driver']/table['Total - Main mode of commuting for the employed labour force aged 15 years and over in private households with a usual place of work or no fixed workplace address - 25% sample data']

table['Non-movers prop'] = table ['Non-movers']/table['Total - Mobility status 1 year ago - 25% sample data']
table['Movers prop'] = table ['Movers']/table['Total - Mobility status 1 year ago - 25% sample data']
```

```python
listofcols = [
 'Total - Occupied private dwellings by structural type of dwelling - 100% data',
 'Single-detached house',
    
 'Total - Marital status for the population aged 15 years and over - 100% data',
 'Married or living common law',
 'Not married and not living common law',

 'Total - Private households by tenure - 25% sample data',
 'Owner',
 'Renter',
    
 'Total - Private households by number of persons per room - 25% sample data',
 'One person or fewer per room',
 'More than 1 person per room',
    
 'Total - Private households by housing suitability - 25% sample data',
 'Suitable',
 'Not suitable',
    
 'Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data',
 'No certificate, diploma or degree',
 'Secondary (high) school diploma or equivalency certificate',
 'Postsecondary certificate, diploma or degree',
    
 'Total - Population aged 15 years and over by Labour force status - 25% sample data',
 'In the labour force',
 'Employed',
 'Unemployed',
 'Not in the labour force',
    
 'Total - Main mode of commuting for the employed labour force aged 15 years and over in private households with a usual place of work or no fixed workplace address - 25% sample data',
 'Car, truck, van - as a driver',
    
 'Total - Mobility status 1 year ago - 25% sample data',
 'Non-movers',
 'Movers']

table.drop(listofcols, axis = 1, inplace = True)
```
