## HA/HSD assignments

```python
# Join DB to PHU assignments to proximity data
merged = pd.merge(left=prox, right=db_phu, how='left', left_on='DBUID', right_on='DBUID')

# Convert population columns to numeric types
merged[['DBPOP']] = merged[['DBPOP']].apply(pd.to_numeric) 

merged.info()
merged.head(5)
```

## Aggregate proximity measures from DB to HSD/HA level

```python
# Create a new column which contains the HSD/HA population number for that db
merged['HSDPOP'] = merged.groupby('HSDA_Name')['DBPOP'].transform('sum')
merged['HAPOP'] = merged.groupby('HA_Name')['DBPOP'].transform('sum')

# Create a new column which is the ratio of the DBPOP/HRPOP
merged['DBtoHSDPOP'] = merged['DBPOP']/merged['HAPOP']
merged['DBtoHAPOP'] = merged['DBPOP']/merged['HSDPOP']

# Inspect the results
merged.info()
# merged['HRPOP'].value_counts()
# merged['HRPOP'].describe()
# # Merged['DBPOP'].value_counts() # 19599 zeros in DBPOP
# merged['DBPOP'].describe()
# # Merged['DBHRPOP'].value_counts() # # 19599 zeros in DBHRPOP
# merged['DBHRPOP'].describe()
```

### 1. Aggregate the proximity measures by taking the sum of population weighted measures for each HR

```python
# Create a list containing prox measures column names
list_prox = ['prox_idx_emp', 'prox_idx_pharma', 'prox_idx_childcare',
             'prox_idx_health', 'prox_idx_grocery', 'prox_idx_educpri',
             'prox_idx_educsec', 'prox_idx_lib', 'prox_idx_parks', 
             'prox_idx_transit']

# Create new weighted proximity columns for each row (ie. DBHRPOP * prox_)
for each in list_prox:
    merged[each + '_HSDwt'] = merged[each] * merged['DBtoHSDPOP']
for each in list_prox:
    merged[each + '_HAwt'] = merged[each] * merged['DBtoHAPOP']
    
# Create a list with weighted prox measures column names
list_prox_HSDwt = [item + '_HSDwt' for item in list_prox]
list_prox_HAwt = [item + '_HSDwt' for item in list_prox]
print(list_prox_HSDwt)
    
# list_prox_wt

# Create new HSD/HA level dfs where weighted proximity values are summed
phu_prox_HSDwt = merged.groupby(['HSDA_Name'])[list_prox_HSDwt].sum()
phu_prox_HAwt = merged.groupby(['HA_Name'])[list_prox_HAwt].sum()
phu_prox_HSDwt.info()
phu_prox_HSDwt.head()
```

### 2. Aggregate the proximity measures by taking the median of the measures for each HR

```python
# Create list of column nammes for the median proximity measures
list_prox_HSDmed = []
list_prox_HAmed = []
for each in list_prox:
    list_prox_HSDmed.append(each + '_HSDmed')
for each in list_prox:
    list_prox_HAmed.append(each + '_HAmed')
```

```python
merged.head()
```

```python
list_proxHSD = ['prox_idx_emp', 'prox_idx_pharma', 'prox_idx_childcare',
             'prox_idx_health', 'prox_idx_grocery', 'prox_idx_educpri',
             'prox_idx_educsec', 'prox_idx_lib', 'prox_idx_parks', 
             'prox_idx_transit', 'HSDPOP']

list_proxHA = ['prox_idx_emp', 'prox_idx_pharma', 'prox_idx_childcare',
             'prox_idx_health', 'prox_idx_grocery', 'prox_idx_educpri',
             'prox_idx_educsec', 'prox_idx_lib', 'prox_idx_parks', 
             'prox_idx_transit', 'HAPOP']

phu_prox_HSDmed = merged.groupby(['HSDA_Name'])[list_proxHSD].median()
phu_prox_HAmed = merged.groupby(['HA_Name'])[list_proxHA].median()

# Rename columns
dictionaryHSD = dict(zip(list_proxHSD, list_prox_HSDmed))
dictionaryHA = dict(zip(list_proxHA, list_prox_HAmed))

# dictionary
phu_prox_HSDmed.rename(columns = dictionaryHSD, inplace=True)
phu_prox_HAmed.rename(columns = dictionaryHA, inplace=True)
phu_prox_HSDmed.info()
phu_prox_HSDmed.head()
```

### 3. Aggregate the amenity dense measure by taking a mean

```python
# Create a df with average amenity dense measure by PHU
phu_amenityHSD = merged.groupby(['HSDA_Name'])['amenity_dense'].mean()
phu_amenityHA = merged.groupby(['HA_Name'])['amenity_dense'].mean()
phu_amenityHSD = phu_amenityHSD.to_frame()
phu_amenityHA = phu_amenityHA.to_frame()
phu_amenityHSD.info()
phu_amenityHSD.head()
```

## Merge and export the aggregated proximity data

```python
# Merge aggregated proximity data (weighted proximity, median proximity and amenity)
# HSD
phu_prox_fullHSD = phu_amenityHSD.merge(phu_prox_HSDwt, how='outer', 
                                  left_on=['HSDA_Name'], 
                                  right_on=['HSDA_Name']).merge(phu_prox_HSDmed, 
                                                                          how='outer', 
                                                                          left_on=['HSDA_Name'], 
                                                                          right_on=['HSDA_Name'])
phu_prox_fullHSD = phu_prox_fullHSD.reset_index()

# HA
phu_prox_fullHA = phu_amenityHA.merge(phu_prox_HAwt, how='outer', 
                                  left_on=['HA_Name'], 
                                  right_on=['HA_Name']).merge(phu_prox_HAmed, 
                                                                          how='outer', 
                                                                          left_on=['HA_Name'], 
                                                                          right_on=['HA_Name'])
phu_prox_fullHA = phu_prox_fullHA.reset_index()

phu_prox_fullHSD.info()
phu_prox_fullHSD.head()
```

```python
# Export the PHU level proximity data
phu_prox_fullHSD.to_csv(r'proximity_data_aggregated_BC_HSD.csv', index = False)
phu_prox_fullHA.to_csv(r'proximity_data_aggregated_BC_HA.csv', index = False)
```
