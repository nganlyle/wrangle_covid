#### Add outbreak status column

```python
def outbreak(x):
    try:
        if int(x) != np.nan:
            return "yes"
    except ValueError:
        return "no"
    
merge['outbreak_status'] = merge['Number of dead'].map(lambda x: outbreak(x))
```

# Cleaning
 * #### As done in Ontario analysis, add binary columns for councils and accreditation

```python
def residents(x):
    if 'Resident' in str(x):
        return 'yes'
    else:
        return 'no'
    
    
def family(x):
    if 'Family' in str(x):
        return 'yes'
    else:
        return 'no'

def accred(x):
    if x == 'Not Accredited':
        return 'no'
    elif x == 'Accredited':
        return 'yes'
    else:
        return 'unknown'
    
merge['residents_council'] = merge.COUNCIL.map(lambda x: residents(x))
merge['family_council'] = merge.COUNCIL.map(lambda x: family(x))
merge['accreditation'] = merge.ACCRED_STATUS.map(lambda x: accred(x))
```

* Fill NA's in number of dead with 0s: 

```python
merge['Number of dead'].fillna(0, inplace = True)
```

```python
# if there is an outbreak and the number of dead = 0, change to unreported 
# (these homes were not on Nora Loreto's reports and the number dead was not reported on gov web, if any)

adjust = merge.query('outbreak_status == "yes" and `Number of dead` == "0"').index.to_list()

merge.loc[adjust]['Number of dead'] = 'Unreported'
```

* Add complete home type names and ownership type cols

```python
def hometype(x):
    if x == "HA":
        return "health_authority"
    elif x == "PNP":
        return "non-profit"
    elif x == "PFP":
        return "for-profit"
    
    
merge['home_type'] = merge['OWNERSHIP'].map(lambda x: hometype(x))

```

* Add years of operation column

current year - `OPEN_DATE`


##### first, fill in some known open_dates

**Refs:**

 * MSA Hospital years operation - https://www.aldergrovestar.com/news/seniors-care-centre-proposed-for-msa-hospital-site/
 
 * Eagle Ridge Manor - Port Moody
https://en.wikipedia.org/wiki/Eagle_Ridge_Hospital

 * Elim village - https://elimvillage.com/2016/04/18/harrison-west-is-open/

 * Hawthorne Seniors Care Community - http://hawthornecare.com/about/


```python
merge.loc[merge['FACILITY_NAME'] == 'Dufferin Care Centre'].OPEN_DATE.replace({'None' :1993}, inplace = True)
merge.loc[merge['FACILITY_NAME'] == 'Cottage and Worthington Pavilions - MSA Hospital'].OPEN_DATE.replace({'None' : 2015}, inplace = True)
merge.loc[merge['FACILITY_NAME'] == 'Eagle Ridge Manor - Port Moody'].OPEN_DATE.replace({'None' :1984}, inplace = True)
merge.loc[merge['FACILITY_NAME'] == 'Elim Village - The Harrison / Harrison West'].OPEN_DATE.replace({'None' : 2016}, inplace = True)
merge.loc[merge['FACILITY_NAME'] == 'Hawthorne Seniors Care Community'].OPEN_DATE.replace({'None' : 1958}, inplace = True)
```

```python
# get today's year
from datetime import datetime
today = datetime.today()
now = today.year

def timeDiff(x):
    try:
        return now - int(x)
    except ValueError:
        pass

merge['years_operating'] = merge.OPEN_DATE.map(timeDiff)
```
