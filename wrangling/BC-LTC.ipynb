### Latitude and Longitude

```python
from geopy import *
import geopandas
```

```python
locator = Nominatim(user_agent="myGeocoder")
```

```python
# reformat address to be one string
address = []
for s,c in zip(merge['STREET_ADDRESS'].to_list(), merge['CITY'].to_list()):
        address.append("{}, {}, British Columbia, Canada".format(s,c))
        
merge['address'] = address
merge.address
```

```python
# try:
#     for p in merge.point:
#         print(p)
# except AttributeError:
#     pass
merge[merge.point.isnull()]
```

<mark> ***Caution: slow script below!*** </mark>

```python
from geopy.extra.rate_limiter import RateLimiter

# 1 - convenient function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

# 2- - create location column
merge['location'] = merge['address'].apply(geocode)

# 3 - create longitude, laatitude and altitude from location column (returns tuple)
merge['point'] = merge['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
merge
```

```python
# 4 - split point column into latitude, longitude and altitude columns
# merge[['latitude', 'longitude', 'altitude']] = pd.DataFrame(merge['point'].tolist(), index=merge.index)

def pt_to_x(pt):
#     print(pt)
    try:
        latitude, longitude, altitude = pt
        return latitude
    except TypeError:
#         print("can't unpack {}".format(type(pt)))
        return pt

def pt_to_y(pt):
    try:
        latitude, longitude, altitude = pt
        return longitude
    except TypeError:
        return pt
    
merge['latitude'] = merge['point'].map(pt_to_x)
merge['longitude'] = merge['point'].map(pt_to_y)
```

# Cleaning
 * #### As done in Ontario analysis, add binary columns for councils and accreditation

```python
def residents(x):
    if 'Resident' in str(x):
        return 'yes'
    else:
        return 'no'
    
    
def family(x):
    if 'Family' in str(x):
        return 'yes'
    else:
        return 'no'

def accred(x):
    if x == 'Not Accredited':
        return 'no'
    elif x == 'Accredited':
        return 'yes'
    else:
        return 'unknown'
    
merge['residents_council'] = merge.COUNCIL.map(lambda x: residents(x))
merge['family_council'] = merge.COUNCIL.map(lambda x: family(x))
merge['accreditation'] = merge.ACCRED_STATUS.map(lambda x: accred(x))
```

* Fill NA's in number of dead with 0s: 

```python
merge['Number of dead'].fillna(0, inplace = True)
```

```python
# if there is an outbreak and the number of dead = 0, change to unreported 
# (these homes were not on Nora Loreto's reports and the number dead was not reported on gov web, if any)

adjust = merge.query('outbreak_status == "yes" and `Number of dead` == "0"').index.to_list()

merge.loc[adjust]['Number of dead'] = 'Unreported'
```

* Add complete home type names and ownership type cols

```python
def hometype(x):
    if x == "HA":
        return "health_authority"
    elif x == "PNP":
        return "non-profit"
    elif x == "PFP":
        return "for-profit"
    
    
merge['home_type'] = merge['OWNERSHIP'].map(lambda x: hometype(x))

```
