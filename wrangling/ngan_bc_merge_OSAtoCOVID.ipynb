```python
# Merge the seniors adv data (ltc) with the covid data
merged = pd.merge(left=ltc, right=covid, 
                  how='left', left_on='cleaned_name', right_on='cleaned_name')
merged.info()
merged.tail()
merged['FACILITY_NAME'].nunique()
merged.tail()
```

```python
# Code to review homes in the db individually/manually
# merged[merged['FACILITY_NAME'].isnull()]
# merged[merged['cleaned_name'].str.contains('worthington', regex=False, na=False)]

merged[merged.duplicated(['FACILITY_NAME'], keep=False)]
```

```python
# Create an outbreak status column
merged['outbreak'] = merged['Facility Name'].isna().apply(lambda x: 0 if x==True else 1)
merged.info()
merged['outbreak'].value_counts()
# merged.head()
```
