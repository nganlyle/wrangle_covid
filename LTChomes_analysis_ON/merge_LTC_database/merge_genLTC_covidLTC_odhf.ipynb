```python
import pandas as pd
import numpy as np

# Set options to show dfs in full
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

## Clean the general LTC data

```python
# Add a column with name in lower case
df_genLTC = pd.read_csv('../data_output/webscrape_LTC_general_database.csv')
df_genLTC['cleaned_name'] = df_genLTC['name'].str.lower()

# Adjust names from general LTC database to match names from COVID LTC database
df_genLTC["cleaned_name"].replace({"vision '74 inc.": "vision nursing home", 
                                   "finlandia hoivakoti nursing home limited": "finlandia hoivakoti nursing home", 
                                   "pinecrest nursing home - bobcaygeon": "pinecrest nursing home (bobcaygeon)", 
                                   "residence saint- louis": "residence saint-louis", 
                                   "st patrick's home": "st. patrick's home",
                                   "heartwood (fka versa-care cornwall)": "heartwood"
                                  }, inplace=True)
df_genLTC.info()
df_genLTC.head()
```

```python
# df_genLTC[df_genLTC['cleaned_name'].str.contains('macassa', regex=False, na=False)]
```

## Merge general LTC and covid LTC data

```python
# Import the covid LTC data

# df_covidLTC = pd.read_csv('merged_ltc_secondScrape.csv')
df_covidLTC = pd.read_csv('../data_output/ltc_outbreaks0731.csv')
df_covidLTC['cleaned_name'] = df_covidLTC['LTC_Home'].str.lower()
df_covidLTC["cleaned_name"].replace({"ã‰lisabeth-bruyã¨re residence": "élisabeth-bruyère residence", 
                                     "macassa  lodge": "macassa lodge"}, inplace=True)
df_covidLTC['outbreak'] = 1

df_covidLTC.info()
df_covidLTC.head()
```

```python
# Check that homes from COVID LTC database are present in the general LTC database
a = set(df_genLTC['cleaned_name'])
b = set(df_covidLTC['cleaned_name'])

def returnNotMatches(a, b):
    return [[x for x in b if x not in a]]

returnNotMatches(a, b) 

# All homes present in the COVID database are present in the general database
# df_covidLTC['cleaned_name'].isin(df_genLTC['cleaned_name']).value_counts()
```

```python
# Merge the general LTC database with the COVID LTC database
df_LTC = pd.merge(left=df_genLTC, right=df_covidLTC, how='outer', left_on='cleaned_name', right_on='cleaned_name')
df_LTC["outbreak"].fillna(0, inplace = True)
```

```python
df_LTC.info()
df_LTC.head()

```

```python
# Check that the number of beds from the general LTC database agree with number of beds from the COVID LTC database
# Subset rows where there is a value for beds from covid dataset
df_LTC_beds = df_LTC[df_LTC.beds_y.notnull()]
df_LTC_beds.info()

# There are 24 homes where the number of beds in the 2 LTC databases do not agree
diff = pd.Series.eq(df_LTC_beds['number_of_beds'],df_LTC_beds['beds_y'])
diff.value_counts()

df_LTC_beds['bed_match'] = np.where((df_LTC_beds['number_of_beds'] == df_LTC_beds['beds_y']), 'match', 'no_match')
df_LTC_beds['bed_match'].value_counts()
```

```python
# Review homes where value for number of beds differ between the databases
df_LTC_beds = df_LTC_beds[df_LTC_beds['bed_match'] == 'no_match']
df_LTC_beds['diff_beds'] = df_LTC_beds['number_of_beds'] - df_LTC_beds['beds_y']
df_LTC_beds[['cleaned_name', 'number_of_beds', 'beds_y', 'diff_beds']]
```
